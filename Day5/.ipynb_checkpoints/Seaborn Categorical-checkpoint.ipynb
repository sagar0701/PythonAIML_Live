```python
#### Categorical Data Plots
import seaborn as sns
# %matplotlib inline

import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

plt.style.use('fivethirtyeight') 

tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.get_dataset_names()
```

```python
tips.sex.value_counts()
```

```python
tips.groupby(['sex','smoker'])['tip'].value_counts()
```

```python
tips.groupby(['sex','smoker']).size().sort_values(ascending=False).to_frame().rename(columns={0:'count'})
```

```python
import numpy as np

a1 = np.array([2,200,20,134,12,14,17,20,45,670])
a2 = np.array([2,20,12,14,17,20])


print(np.mean(a1),np.mean(a2),
      '\n', np.var(a1), np.var(a2),\
      '\n', np.std(a1),np.std(a2))

print(np.median(a1))
```

```python
tips.describe()
# min bill -3, max bill - 350, std - 200, mean- 130, 75%- 350 , 25%- 50 , median - 80

fig = plt.figure(figsize = (7,5))
sns.heatmap(tips.describe(), annot = True, fmt = '0.4f',  cmap="winter")
plt.show()
```

```python
tips.corr()  #[-1, +1], Pearson correaltion 

# expenditure  -- Saving - [-1, +1]  
# negative corr .  expenditure increase --- saving- decr VS expenditure decrease, Saving - increase -1
# Positive coorealtipon-- Salary - Benifits - positive - +1
# no realtion - coorelation -- Salary and time taken to reach ofc  - aroud 0
```

```python
#https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
fig = plt.figure(figsize = (7,5))
sns.heatmap(tips.corr(), annot = True, fmt = '0.2f',  cmap="gist_rainbow")
```

```python
tips['total_bill'].max()
```

```python
# Bar Plot
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.max)
```

```python
tips.groupby('day').time.value_counts()
```

```python
# Count plot
sns.countplot(x='sex',data=tips, hue='size')
```

```python
# Box pLot
sns.boxplot(x = 'total_bill', data = tips)
```

```python
sns.boxplot(x="time", y="total_bill", data=tips, palette='prism', hue='sex')
```

```python
sns.violinplot(x="sex", 
               y="total_bill", 
               data=tips,
               hue='day',
               palette='prism')
```

```python
# Strip Plot
sns.stripplot(x="day", y="total_bill", data=tips,
              jitter=False,
              hue='sex',
              palette='gist_ncar')
```

```python
# Swarm plot
sns.swarmplot(x="day", y="total_bill",
              hue='sex',
              data=tips,
              palette="gist_ncar",
              split=True)
```
