```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('train.csv')
df.head()
```

```python
df.info()
```

```python
# If for any column(s), the variance is equal to zero,then you need to remove those variable(s).
for col in df.columns:
    if df[col].dtype!=object:
        if(df.var()[col]==0.0):
            print(col,df.var()[col])
            df.drop([col],axis=1,inplace=True)           
```

```python
# Check for null and unique values for test and train sets
for col in df:
      print(col +': '+ np.str(df[col].isna().sum()))
```

```python
for col in df:
    print(col,df[col].unique())
```

```python
df.info()
```

```python
# Apply label encoder.
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
l_train=['X0','X1','X2','X3','X4','X5','X6','X8']
for col in l_train:
    print(col)
    le.fit(df[col].values)
    df[col]=le.transform(df[col].values)
```

```python
#Perform dimensionality reduction.
from sklearn.decomposition import PCA
pca=PCA(n_components=14)
df_reduced = pca.fit_transform(df.drop(['y'],axis=1))
df_reduced
```

```python
df_reduced.shape
```

```python
# Predict your test_df values using xgboost
from sklearn import model_selection
from sklearn.ensemble import AdaBoostRegressor

X = df_reduced[:,0:14]
Y = df_reduced[:,1]
num_trees = 30
```

```python
Y
```

```python
kfold = model_selection.KFold(n_splits=10)
model = AdaBoostRegressor(n_estimators=num_trees)
results = model_selection.cross_val_score(model, X, Y, cv=kfold)
print(results.mean())
```

```python
from xgboost import XGBRegressor

seed = 7
num_trees = 30
kfold = model_selection.KFold(n_splits=10)
model = XGBRegressor(n_estimators=num_trees, random_state=seed, eval_metric='logloss', learning_rate=0.001)
results = model_selection.cross_val_score(model, X, Y, cv=kfold)
print(results.mean())
```
