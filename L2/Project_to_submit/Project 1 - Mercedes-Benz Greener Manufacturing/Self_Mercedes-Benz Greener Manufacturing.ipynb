```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('train.csv')
df.head()
```

```python
df_test=pd.read_csv('test.csv')
df_test.head()
```

```python
df.info()
```

```python
# If for any column(s), the variance is equal to zero,then you need to remove those variable(s).
for col in df.columns:
    if df[col].dtype!=object:
        if(df.var()[col]==0.0):
            print(col,df.var()[col])
            df.drop([col],axis=1,inplace=True)
            
for col in df_test.columns:
    if df_test[col].dtype!=object:
        if(df_test.var()[col]==0.0):
            print(col,df_test.var()[col])
            df_test.drop([col],axis=1,inplace=True)            
```

```python
# Check for null and unique values for test and train sets
for col in df:
      print(col +': '+ np.str(df[col].isna().sum()))
     
        
for col in df_test:
      print(col +': '+ np.str(df_test[col].isna().sum()))
        
```

```python
for col in df:
    print(col,df[col].unique())
    
    
for col in df_test:
    print(col,df_test[col].unique())
```

```python
# Apply label encoder.
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df[['X0','X1','X2','X3','X4','X5','X6','X8']]=df[['X0','X1','X2','X3','X4','X5','X6','X8']].apply(le.fit_transform)

```

```python
df_test[['X0','X1','X2','X3','X4','X5','X6','X8']]=df_test[['X0','X1','X2','X3','X4','X5','X6','X8']].apply(le.fit_transform)
```
