```python
import pandas as pd          
import numpy as np          
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings("ignore")

%matplotlib inline

from statsmodels.tsa.arima_model import ARIMA, ARMA
from statsmodels.tsa.ar_model import AR
```

```python
df = pd.read_csv('monthly-beer-production-in-austr.csv')
print(df.info())
df.head()
```

```python
df['Month'] = pd.to_datetime(df['Month'])

print(df.info())

indexed_df = df.set_index('Month')
print(indexed_df.head())
```

```python
Module_making.ts(indexed_df)
```

```python
indexed_df_log = np.log(indexed_df)
print(indexed_df_log.head())
```

```python
Module_making.ts(indexed_df_log) 
```

```python
x2 = indexed_df[:130] # train data
y2 = indexed_df[130:] # test data

Module_making.acf_pacf_plots(x2)
```

```python
# ARIMA model
model  = ARIMA(x2, order = (12,1,0))  # (p , d(differenced) , q) values , we can use different values to make model good or worse

results_ar_2 = model.fit()

pred_2 = results_ar_2.predict(start = y2.index[0], end = y2.index[-1],typ = 'levels')
```

```python
pred_2.head()
```

```python
df_pred_2 = pd.DataFrame({'Actual': y2['Monthly beer production in Australia'], 'Predicted':pred_2})

fig = plt.figure(figsize=(13,7))

plt.plot(df_pred_2['Actual'],color = 'red', label = 'orignal')

plt.plot(df_pred_2['Predicted'],color = 'black',label = 'predicted')

plt.plot(x2['Monthly beer production in Australia'], color = 'yellow', label = 'train')

plt.legend();

# To check results
Module_making.evaluate_forecast(y2['Monthly beer production in Australia'], pred_2)
```

```python
forecast = results_ar_2.forecast(steps = 4)

print(' # Forecasted Value: ', forecast[0]) # Forecasted Value
```
