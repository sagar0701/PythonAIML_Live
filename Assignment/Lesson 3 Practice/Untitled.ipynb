```python
df_salary=pd.read_csv('Salaries.csv')
df_salary.head()
```

```python
sns.barplot(df_salary['Year'],df_salary['TotalPayBenefits'])
```

```python
#!  How much total salary cost has increased from year 2011 to 2014
ck=df_salary.groupby('Year')['TotalPayBenefits'].sum().to_frame()
ck['shift']=ck.shift(periods=1)
ck['%diff']=(ck['shift']/ck['TotalPayBenefits'])*100
ck
```
